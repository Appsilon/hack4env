```python
import pandas as pd
```

```python
df = pd.read_csv("jsondata_nona.csv")
df_brands = df[df.result_string.str.contains("brands.")]
df_brands
```

```python
df_brands["brands"] = df_brands.result_string.str.extract("brands\.([^\s]*)\s")
df_brands
```

```python
df_brands.brands.value_counts()
```

# Download the images

```python
import urllib.request

counter = 6400
for url in df_brands.filename[6400:]:
    filename = url.split(".com/")[1].replace("/","@")
    try:
        urllib.request.urlretrieve(url, f'/home/jupyter/brand_images/{filename}')
    except:
        print("error, going further...")
        continue
    counter += 1
    if counter % 100 == 0:
        print(f"file {counter} downloaded")
```
